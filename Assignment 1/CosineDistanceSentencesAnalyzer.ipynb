## Найти два предложения, которые ближе всего по смыслу к расположенному в самой первой строке. В качестве меры близости по смыслу использовать косинусное расстояние.

```python
import numpy
import string
import re
import pandas
```

```python
file_obj = open('sentences.txt', 'r')
```

```python
linesList = list(file_obj)
```

```python
linesArray = []
wordsList = []
for line in linesList:
    loweredLine = line.strip().lower()
    linesArray.append(loweredLine)
    #word = loweredLine.split(" ")
    words = re.split('[^a-z]', loweredLine)
    #print(word)
    words = filter(lambda x:x.strip(), words)
    wordsList.append(words)
```

### Create dictionary of all words in all sentences

```python
wordsFrequencies = dict()

for currentSentence in wordsList:
    for word in currentSentence:
        wordsFrequencies[word] = 0

wordsFrequencies
```

```python
for key in wordsFrequencies:
    for currentSentence in wordsList:
        for word in currentSentence:
            if(word == key):
                wordsFrequencies[key] += 1 
```

```python
wordsFrequencies
```

### Creating list of dictionaries

```python
listOfDicts = []
for currentSentence in wordsList: 
    currentSentenceDict = dict()
    for word in currentSentence:
        currentSentenceDict[word] = 0
    listOfDicts.append(currentSentenceDict)
```

```python
for i in range(len(wordsList)):
    currentDict = listOfDicts[i]
    currentSentence = wordsList[i]
    for word in currentSentence:
        for key in currentDict:
            if(word == key):
                currentDict[key]+=1
```

```python
listOfDicts
```

### Create entry matrix 

```python
entryMatrix = numpy.zeros([len(wordsList), len(wordsFrequencies)])
```

```python
entryMatrix.shape
```

```python
keysList = []
for key in wordsFrequencies: 
    keysList.append(key)
```

```python
for i in range(len(wordsList)):
    currentDict = listOfDicts[i]
    #print(currentDict)
    for j in range(len(keysList)):
        for key in currentDict:
            if key == keysList[j]:
                entryMatrix[i][j] = currentDict.get(key)
                #print(currentDict.get(key))
```

```python
entryMatrix
```

```python
frame = pandas.DataFrame(entryMatrix, columns = keysList) 
```

```python
import scipy.spatial.distance
distances = []
for row in entryMatrix:
    distances.append(scipy.spatial.distance.cosine(entryMatrix[0,:], row))
```
